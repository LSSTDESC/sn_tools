```python
import numpy as np
import os
import pylab as plt
```

```python
rate = 'JLA'
prefix = os.getenv('SN_UTILS_DIR')+'/input/Dist_X1_Color_'+rate+'_'
suffix = '.txt'
# names=['x1','c','weight_x1','weight_c','weight_tot']
dtype = np.dtype([('X1', np.float), 
                ('Color', np.float),
                ('weight_X1', np.float), 
                ('weight_Color', np.float),
                ('weight', np.float)])
x1_color = {}
for val in ['low_z', 'high_z']:
    x1_color[val] = np.loadtxt(prefix+val+suffix, dtype=dtype)
    print(x1_color[val][['Color','weight']])
```

```python
def Plot(what,x1_color,var):
    fig, ax = plt.subplots(ncols=1,nrows=1)
    x1_color_disp = x1_color[what]
    ax.hist(x1_color_disp[var],weights=x1_color_disp['weight'],bins=len(x1_color_disp[var]))
    ax.set_xlabel(var)
    ax.set_ylabel('Number of Entries')
```

```python
#%matplotlib
Plot('low_z',x1_color,'Color')
Plot('low_z',x1_color,'X1')
```

```python
Plot('high_z',x1_color,'Color')
Plot('high_z',x1_color,'X1')
```
